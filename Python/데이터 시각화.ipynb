```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.subplot(221), plt.bar(x, values, color='y'),plt.xticks(x, years)
plt.subplot(222), plt.bar(x, values, color='dodgerblue'),plt.xticks(x, years)
plt.subplot(223), plt.bar(x, values, color='C2'),plt.xticks(x, years)
plt.subplot(224), plt.bar(x, values, color='#e35f62'),plt.xticks(x, years)
plt.show()
```

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

#plt.pie(ratio, labels=labels, autopct='%.1f%%') #autopct는 부채꼴 안에 
# 표시될 숫자의 형식을 지정합니다. 소수점 한자리까지 표시하도록 설정
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, 
        counterclock=False)
# startangle는 부채꼴이 그려지는 시작 각도
# 디폴트는 0도 (양의 방향 x축)로 설정
# counterclock=False로 설정하면 시계 방향 순서로 부채꼴 영역이 표시
plt.show()
```

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]  # explode는 부채꼴이 파이 차트의 
#중심에서 벗어나는 정도를 설정
#'Banana’와 ‘Grapes’영역에 대해서 반지름의 10% 만큼 벗어나도록 설정
#plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260,
# counterclock=False, explode=explode)

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, 
        counterclock=False, explode=explode, shadow=True) #그림자 설정
plt.show()
```

```python
atio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]
#colors = ['silver', 'gold', 'whitesmoke', 'lightgray'] #색 지정
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0'] 
#Hex code를 이용해서 더욱 다양한 색상을 지정

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, 
        counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.subplot(2,2,1)
plt.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
plt.subplot(2,2,2)
plt.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
plt.subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(), 'k--')
plt.show()
```
