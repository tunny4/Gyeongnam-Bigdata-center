```python
import pandas as pd

df=pd.read_csv("d:/데이터/소상공인시장진흥공단_상가(상권)정보_경남_202309.csv", index_col=False, encoding="UTF-8", low_memory=False)
```

```python
df.loc[df['도로명주소'].str.contains('창원'), '도로명주소']='창원'
df.loc[df['도로명주소'].str.contains('진주'), '도로명주소']='진주'
df.loc[df['도로명주소'].str.contains('김해'), '도로명주소']='김해'

df.rename(columns={'상권업종소분류명':'업종'}, inplace=True)
df.rename(columns={'도로명주소':'주소지'}, inplace=True)

df = df[['상가업소번호', '업종', '주소지', '위도', '경도']]
df.head()
```

```python
df
```

```python
top13 = ['기타 의류 소매업', '슈퍼마켓', '미용실', '편의점', '요리 주점',
 '일반 유흥 주점', '카페', '부동산 중개/대리업', '여성 의류 소매업', '김밥/만두/분식',
 '치킨', '백반/한정식', '돼지고기 구이/찜']

df = df.loc[df['업종'].isin(top13)]
df
```

```python
import matplotlib.pyplot as plt
import matplotlib as mat
import seaborn as sns

mat.rcParams['font.family'] = 'NamumGothic'

fig, ax1 = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(12, 5)

plt.ylim([0, 14000]) # y축 값의 범위
plt.xticks(rotation = 45)
sns.countplot(df['업종'], order=df['업종'].value_counts().index)  # 경남전체
```
